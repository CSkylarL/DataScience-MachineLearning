## Introduction

The [Iris dataset](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html) is one of the most famous datasets in the field of machine learning. It was introduced by the British statistician and biologist Ronald Fisher in his 1936 paper "The use of multiple measurements in taxonomic problems."

The dataset consists of 150 samples from each of three species of Iris flowers (Iris setosa, Iris virginica, and Iris versicolor). Four features were measured from each sample: the lengths and the widths of the sepals and petals.

Here's a brief overview of the dataset:

- **Number of Instances**: 150 (50 in each of three classes)
- **Number of Attributes**: 4 numeric, predictive attributes and the class
- **Attribute Information**:
  1. sepal length in cm
  2. sepal width in cm
  3. petal length in cm
  4. petal width in cm
  5. class:
     - Iris Setosa
     - Iris Versicolour
     - Iris Virginica

The dataset is often used in data mining, classification and clustering examples and to test algorithms. It's available in the [scikit-learn library](https://scikit-learn.org/stable/), a popular machine learning library in Python.



## Load data

[load_iris()](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html) function from the scikit-learn library is used to load the dataset. It has the following characteristics:

Again, each instance in the dataset represents a flower. The four features are numerical values representing the length and width of the sepals and petals. The target variable is a categorical variable with three possible values representing the three flower species.

````python
from sklearn.datasets import load_iris
# Load the dataset
data = load_iris()

# Display the shape of the data, target, and number of classes
print("Data shape:", data.data.shape)
print("Target shape:", data.target.shape)
print("Number of classes:", len(data.target_names))
print("Classes:", data.target_names)
````

I also combined the data and target into a DataFrame and saved the data to [iris.csv](./iris.csv) file in the same directory as this notebook, so that you can load the data from the numpy files directly.

    ```python
    import pandas as pd

    # Combine the data and target into a DataFrame
    df = pd.DataFrame(data.data, columns=data.feature_names)
    df['target'] = data.target

    # Save the DataFrame to a .csv file
    df.to_csv('iris.csv', index=False)
    ```
    
You can load the data using below code:

````python
import pandas as pd

# Read the .csv file back into a DataFrame
df = pd.read_csv('iris.csv')

# Display the first few rows of the DataFrame
df
````
