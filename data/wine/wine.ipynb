## Introduction

The [Wine dataset](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_wine.html#sklearn.datasets.load_wine) is a classic and very easy multi-class classification dataset provided by  [UCI Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets/wine). This data has been included in the Scikit-Learn datasets module for convenience.

There are 178 samples in this dataset. Each sample is described by 13 feature variables, which are different physicochemical tests that are conducted on a wine. These tests include attributes like alcohol content, malic acid content, magnesium content, etc.

Each sample in the dataset is described by 13 numeric, predictive attributes:

1. Alcohol
2. Malic acid
3. Ash
4. Alcalinity of ash
5. Magnesium
6. Total phenols
7. Flavanoids
8. Nonflavanoid phenols
9. Proanthocyanins
10. Color intensity
11. Hue
12. OD280/OD315 of diluted wines
13. Proline

All attributes are continuous, with no missing values, which makes it a good starting point for beginners in machine learning to practice classification algorithms.


The task on this dataset is to predict the origin of the wine, which is a categorical variable with three possible values, corresponding to the three different cultivars from which the wines originate:

- Class 0: Cultivar 1
- Class 1: Cultivar 2
- Class 2: Cultivar 3

The classes are not balanced, meaning there are unequal numbers of instances in each class.

This dataset is a good starting point for beginners in machine learning to practice classification algorithms, especially for multi-class classification problems.


## Load data

[load_wine()](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_wine.html#sklearn.datasets.load_wine) function from the scikit-learn library is used to load the dataset. It has the following characteristics:

- Number of instances (samples): 178
- Number of attributes (features): 13
- Number of classes: 3
- Class names: ['class_0', 'class_1', 'class_2']

Each instance in the dataset represents a wine. The attributes are the results of a chemical analysis of wines grown in the same region in Italy but derived from three different cultivars. The analysis determined the quantities of 13 constituents found in each of the three types of wines.

The target variable is a categorical variable with three possible values, corresponding to the three different cultivars from which the wines originate.



````python
from sklearn.datasets import load_wine
# Load the dataset
data = load_wine()
# Display the shape of the data, target, and number of classes
print("Data shape:", data.data.shape)
print("Target shape:", data.target.shape)
print("Number of classes:", len(data.target_names))
print("Classes:", data.target_names)
````

I also combined the data and target into a DataFrame and saved the data to [wine.csv](./wine.csv) file in the same directory as this notebook, so that you can load the data from the numpy files directly.

    ```python
    import pandas as pd

    # Combine the data and target into a DataFrame
    df = pd.DataFrame(data.data, columns=data.feature_names)
    df['target'] = data.target

    # Save the DataFrame to a .csv file
    df.to_csv('wine.csv', index=False)
    ```
    
You can load the data using below code:

````python
import pandas as pd

# Read the .csv file back into a DataFrame
df = pd.read_csv('wine.csv')

# Display the first few rows of the DataFrame
df
````
