## Introduction

The [Breast Cancer dataset](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_breast_cancer.html) available in the scikit-learn library is a classification dataset. It is a copy of the [UCI ML Breast Cancer Wisconsin (Diagnostic) dataset](https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+(Diagnostic)). 

The dataset has 569 instances or samples, each with 30 numeric, predictive features or attributes and a target, which can be either 'malignant' (cancerous) or 'benign' (non-cancerous). The features are computed from a digitized image of a fine needle aspirate (FNA) of a breast mass. They describe characteristics of the cell nuclei present in the image.

The dataset's features are:

- radius (mean of distances from center to points on the perimeter)
- texture (standard deviation of gray-scale values)
- perimeter
- area
- smoothness (local variation in radius lengths)
- compactness (perimeter^2 / area - 1.0)
- concavity (severity of concave portions of the contour)
- concave points (number of concave portions of the contour)
- symmetry 
- fractal dimension ("coastline approximation" - 1)

The mean, standard error, and "worst" or largest (mean of the three largest values) of these features were computed for each image, resulting in 30 features. 

This dataset is useful for binary classification tasks, specifically for developing and testing machine learning models for cancer diagnosis.


## Load data

[load_breast_cancer()](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_breast_cancer.html) function from the scikit-learn library is used to load the dataset. It has the following characteristics:

- Number of instances (samples): 569
- Number of attributes (features): 30
- Number of classes: 2
- Class names: 'malignant' and 'benign'

Each instance in the dataset represents a patient's record, with 30 features derived from a digitized image of a fine needle aspirate (FNA) of a breast mass. The features capture important characteristics of the cell nuclei present in the image. 

The target variable is a binary classification - either 'malignant' (indicating the presence of cancer cells) or 'benign' (indicating the absence of cancer cells).



````python
from sklearn.datasets import load_breast_cancer
# Load the dataset
data = load_breast_cancer()

# Display the shape of the data, target, and number of classes
print("Data shape:", data.data.shape)
print("Target shape:", data.target.shape)
print("Number of classes:", len(data.target_names))
print("Classes:", data.target_names)
````

I also combined the data and target into a DataFrame and saved the data to [breast_cancer.csv](./breast_cancer.csv) file in the same directory as this notebook, so that you can load the data from the numpy files directly.

    ```python
    import pandas as pd

    # Combine the data and target into a DataFrame
    df = pd.DataFrame(data.data, columns=data.feature_names)
    df['target'] = data.target

    # Save the DataFrame to a .csv file
    df.to_csv('breast_cancer.csv', index=False)
    ```
    
You can load the data using below code:

````python
import pandas as pd

# Read the .csv file back into a DataFrame
df = pd.read_csv('breast_cancer.csv')

# Display the first few rows of the DataFrame
df
````
