# Introduction

The [Digits dataset](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_digits.html#sklearn.datasets.load_digits) is a sample dataset provided by Scikit-learn, ideally used in pattern recognition. It is a copy of the test set of the [UCI ML hand-written digits datasets](https://archive.ics.uci.edu/ml/datasets/Optical+Recognition+of+Handwritten+Digits). 

The dataset is a preprocessed set of images of handwritten digits, originating from a database created by the National Institute of Standards and Technology. It is a small but substantial dataset, consisting of 1,797 8x8 images. Each image is of a hand-written digit, making it a useful tool for training and testing in the realm of machine learning.

In the scikit-learn version of the dataset, these images are flattened into 64-element feature vectors, representing the grayscale values of each pixel in the image. The target field is an integer from 0 to 9, corresponding to the actual digit the handwritten image represents.

This dataset has been processed and curated separately. Due to its smaller size and lower resolution, it is particularly well-suited for teaching purposes, demonstrating machine learning techniques on real-world data without necessitating substantial computational resources.

- **Data Set Characteristics:** 

    - Number of Instances: 1797
    - Number of Attributes: 64
    - Attribute Information: 8x8 image of integer pixels in the range 0-16.
    - Missing Attribute Values: None

- **Target Set Characteristics:**

    - Number of Classes: 10
    - Attribute Information: Class (0-9)
    - Missing Attribute Values: None
    - Class Distribution: uniform



## Load data

[load_digits()](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_digits.html) function from the scikit-learn library is used to load the dataset. It has the following characteristics:

The data set contains images of hand-written digitsã€‚ Each instance in the dataset represents a 8x8 image of a digit. The target variable contains 10 classes where each class refers to a digit.


````python
from sklearn.datasets import load_digits
# Load the dataset
data = load_digits()

# Display the shape of the data, target, and number of classes
print("Data shape:", data.data.shape)
print("Target shape:", data.target.shape)
print("Number of classes:", len(data.target_names))
print("Classes:", data.target_names)

# X is the data, y is the target
X, y = data.data, data.target 
````

I also saved the data to using numpy in the same directory as this notebook, so that you can load the data from the numpy files directly.

    ```python
    # Save the data
    import numpy as np
    np.save('X.npy', X)
    np.save('y.npy', y)
    ```
    
You can load the data using below code:

````python
import numpy as np

# Load the data
X = np.load('X.npy')
y = np.load('y.npy')
````

## Data target distribution

Here is the target distribution of the dataset:

````python
import seaborn as sns
import matplotlib.pyplot as plt
# Visualize the distribution of the target variable using a bar plot:
sns.countplot(x=y)
plt.show()
````

## Data Visualization
Let's visualize the first 50 images from the training dataset and display the class name below each image. You can run the code below to see the images.

````python
# Visualize the first 50 images in the train set and their labels
n_rows = 5
n_cols = 10
plt.figure(figsize=(n_cols * 1.2, n_rows * 1.2))
for row in range(n_rows):
    for col in range(n_cols):
        index = n_cols * row + col
        plt.subplot(n_rows, n_cols, index + 1)
        plt.imshow(X[index].reshape(8, 8), cmap="binary", interpolation="nearest")
        plt.axis('off')
        plt.title(y[index], fontsize=12)
plt.subplots_adjust(wspace=0.2, hspace=0.5)

plt.show()
````
